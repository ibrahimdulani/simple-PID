```python
import time
import matplotlib.pyplot as plt
import PID
import numpy as np
```

### A simple PID controller for car speed

```python
v0 = 25

ts=0.1

time_steps = np.arange(0,10,ts)

target=55

pid = PID.PID(2,1,1,target,ts)

v_last = 25

vs=[25]

for t in range(10):
    
    a = pid.compute(v_last)
    
    v=v_last + a*ts
    
    vs.append(v)
    
    v_last = v 
```

```python
tt = np.arange(0,11*ts,ts)

plt.figure(figsize=(10,6))

plt.plot(tt,vs)
plt.plot(tt, target*np.ones(11), label='target')
plt.xlabel('time')
plt.ylabel('speed')
plt.legend()
plt.show()
```
